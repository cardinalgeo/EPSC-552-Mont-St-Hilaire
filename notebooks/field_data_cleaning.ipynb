```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('./data/interim/field_data_coarse_clean.xlsx')

data.head()
```

```python
## make all characters lowercase for a subset of fields
lower_case_fields = ["Weather", "Type", "Type Sed", "Land use", "Disturbance", "Comments", "lineNS", "lineEW"]
for field in lower_case_fields: 
    data[field] = data[field].str.lower()        
```

```python
## remove unnecessary data from "Duplicate" field
data["Duplicate"] = data["Duplicate"].replace("No", np.NaN)
data["Duplicate"] = data["Duplicate"].replace("Yes", np.NaN)
```

```python
## Remove NS and EW prefixes from "lineNS" and "lineEW" fields respectively; remove leading 0
data["lineNS"] = data["lineNS"].map(lambda x: str(x).lstrip("ns"))
data["lineNS"] = data["lineNS"].map(lambda x: str(x).lstrip("0"))

data["lineEW"] = data["lineEW"].map(lambda x: str(x).lstrip("ew"))
data["lineEW"] = data["lineEW"].map(lambda x: str(x).lstrip("0"))
```

```python
data.head()
```

```python
## Replace variations in "Land Use" field
data["Land use"] = data["Land use"].replace({"forested": "forest"})
```

```python
data.to_excel('./data/interim/field_data_fine_clean.xlsx', 'field_data')
```
